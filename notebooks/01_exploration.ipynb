```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return pd.DataFrame(data)
```

```python
fr_data_path = '../data/questions_fr.json'
en_data_path = '../data/questions_en.json'
```

```python
fr_data = load_data(fr_data_path)
en_data = load_data(en_data_path)
```

```python
print("Données en français :")
print(fr_data.head())
print("\nDonnées en anglais :")
print(en_data.head())
```

```python
def display_statistics(data, language):
    print(f"\nStatistiques des données en {language} :")
    print(data.describe(include='all'))

display_statistics(fr_data, 'français')
display_statistics(en_data, 'anglais')
```

```python
def plot_question_length_distribution(data, language):
    data['question_length'] = data['question'].apply(len)
    plt.figure(figsize=(10, 6))
    sns.histplot(data['question_length'], bins=50, kde=True, color='skyblue')
    plt.title(f'Distribution des longueurs des questions en {language}')
    plt.xlabel('Longueur des questions')
    plt.ylabel('Fréquence')
    plt.show()
```

```python
plot_question_length_distribution(fr_data, 'français')
plot_question_length_distribution(en_data, 'anglais')
```

```python
def plot_response_length_distribution(data, language):
    data['response_length'] = data['réponse'].apply(len)
    plt.figure(figsize=(10, 6))
    sns.histplot(data['response_length'], bins=50, kde=True, color='lightgreen')
    plt.title(f'Distribution des longueurs des réponses en {language}')
    plt.xlabel('Longueur des réponses')
    plt.ylabel('Fréquence')
    plt.show()

plot_response_length_distribution(fr_data, 'français')
plot_response_length_distribution(en_data, 'anglais')
```
